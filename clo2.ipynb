```python
# libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, ttest_1samp
from sklearn.linear_model import LinearRegression
```

```python
# recognize the dataset

data = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
# descriptive function

def descriptive_stats(data, column):
    return {
        "Mean": data[column].mean(),
        "Median": data[column].median(),
        "Std": data[column].std(),
        "Min": data[column].min(),
        "Max": data[column].max()
    }
```

```python
# random sample of the data

sample = data.sample(n=150, random_state=1)
```

```python
# Systematic sample

sample_sys = data.iloc[::10]
```

```python
# Stats
print(descriptive_stats(sample, "Exam_Score"))
print(descriptive_stats(sample_sys, "Exam_Score"))
```

```python
# Scatter
plt.scatter(data["Hours_Studied"], data["Exam_Score"])
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

```python
# Histogram
plt.hist(data["Exam_Score"])
plt.show()
```

```python
#Boxplot

plt.boxplot(data["Exam_Score"])
plt.show()
```

```python
# Heatmap
sns.heatmap(data.corr(numeric_only=True), annot=True)
plt.show()
```

```python
# Correlation
print(pearsonr(data["Hours_Studied"], data["Exam_Score"]))
print(spearmanr(data["Hours_Studied"], data["Exam_Score"]))
```

```python
# T-test
print(ttest_1samp(sample["Exam_Score"], data["Exam_Score"].mean()))
```

```python
# Regression
model = LinearRegression()
model.fit(data[["Hours_Studied"]], data["Exam_Score"])
```
